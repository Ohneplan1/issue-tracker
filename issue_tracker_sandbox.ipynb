```python
from sqlite3 import connect
```

```python
connection = connect("issue_tracker.db")
```

```python
def initialize_database():
    connection = connect("issue_tracker.db")
    sql = """CREATE TABLE issue (issue_id integer PRIMARY KEY AUTOINCREMENT,
name TEXT, 
description TEXT, 
due_date DATE, 
priority INTEGER);
"""

    cursor = connection.execute(sql)
    connection.commit()
    cursor.close()
    connection.close()
```

```python
def add_issue(name, description, due_date, priority):
    connection = connect("issue_tracker.db")
    sql = f" INSERT INTO issue (name, description, due_date, priority) values ('{name}', '{description}', '{due_date}', {priority});"
    
    cursor = connection.execute(sql)
    connection.commit()
    cursor.close()
    connection.close()
```

```python
add_issue(
    name="Staubsaugen", 
    description="Putze die Wohnung",
    due_date="2026-07-30",
    priority=1
)
```
